```python
# Определяем успешные статусы
successful_statuses = ["первичный контакт", "собеседование", "тел. интервью", "предложение о работе", "выход на работу"]

# Подсчитываем количество каждого статуса
status_distribution = data["Информация об этапе воронки найма.Информация о состоянии.name"].value_counts()

# Отбираем только успешные статусы
successful_status_distribution = status_distribution[status_distribution.index.isin(successful_statuses)]

# Вывод распределения успешных статусов
print("Распределение успешных статусов:")
print(successful_status_distribution)
```

```python
# Список успешных статусов
successful_statuses = ["первичный контакт", "собеседование", "тел. интервью", "предложение о работе", "выход на работу"]

# Обновляем Target с учётом географического соответствия
def define_target(row):
    if row["Информация об этапе воронки найма.Информация о состоянии.name"] in successful_statuses:
        if (
            row["Информация об этапе воронки найма.Информация о состоянии.name"] == "первичный контакт"
            and row["Географическое соответствие"] == 0
        ):
            return 0  # Исключаем первичный контакт без географического соответствия
        return 1
    else:
        return 0

# Применяем функцию для определения Target
data["Target"] = data.apply(define_target, axis=1)

# Проверяем распределение нового Target
new_distribution = data["Target"].value_counts()

# Выводим распределение
print("Распределение нового Target:")
print(new_distribution)
```

```python
correlation = data["Общее количество сообщений"].corr(data["Target"])
print(f"Корреляция между количеством сообщений и Target: {correlation}")
```
