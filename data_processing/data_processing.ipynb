```python
import pandas as pd
```

```python
data = pd.read_excel('/Users/rii_beltz/Desktop/candidate_classifier/data/interim/updated_dataset.xlsx')
```

```python
data['Источник отклика'].value_counts()
```

```python
data_new = data.drop(columns=['Источник отклика'])
```

```python
pd.set_option('display.max_columns', None) 
print(data_new.columns.to_list())
```

```python
city_to_id = dict(zip(data_new['Город проживания'], data_new['Город проживания.id']))

data_new['Город ID'] = data_new['Город'].map(city_to_id)

print(data_new[['Город', 'Город проживания', 'Город ID']].tail(100))
```

```python
data_new = data_new.drop(columns=['Город'])
```

```python
data_new.head()
```

```python
cols = ['Город ID', 'Регион вакансии'] + [col for col in data_new.columns if col not in ['Город ID', 'Регион вакансии']]
data_new = data_new[cols]
```

```python
# словарь для регионов
region_mapping = {region: idx for idx, region in enumerate(data_new['Регион'].unique())}

data_new['Регион_id'] = data_new['Регион'].map(region_mapping)

region_vacancy_mapping = {region: idx for idx, region in enumerate(data_new['Регион вакансии'].unique())}
data_new['Регион вакансии_id'] = data_new['Регион вакансии'].map(region_vacancy_mapping)
```

кодируем столбцы, удаляем ненужные

```python
data_new.get_dummies(columns=['Текущее состояние отклика.Идентификатор статуса'], inplace=True)

data_new.head()
```

```python
data_new.drop(columns=['Текущее состояние отклика.Название статуса'], inplace=True)
```

```python
data_new.drop(columns=['Желаемая должность'], inplace=True)
```

```python
data_new.drop(columns=['Город проживания'], inplace=True)
data_new.drop(columns=['Регион'], inplace=True)
```

```python
data_new = pd.get_dummies(data_new, columns=['Пол соискателя.id'])
data_new.head()
```

```python
data_new.drop(columns=['Пол соискателя.name'], inplace=True)
```

```python
data_new.drop(columns=['Язык, на котором составлено резюме.id'], inplace=True)
data_new.drop(columns=['Язык, на котором составлено резюме.name'], inplace=True)
```

```python
data_new['Образование соискателя.level.id'] = data_new['Образование соискателя.level.id'].replace(
    {'candidate': 'advanced', 'doctor': 'advanced'}
)
education_mapping = {
    'secondary': 1,
    'special_secondary': 2,
    'unfinished_higher': 3,
    'bachelor': 4,
    'higher': 5,
    'master': 6,
    'advanced': 7
}

data_new['Образование_категории'] = data_new['Образование соискателя.level.id'].map(education_mapping)

data_new.head()
```

```python
data_new.drop(columns=['Образование соискателя.level.name'], inplace=True)
```

```python
data_new = pd.get_dummies(data_new, columns=['Подходящих тип занятости.id'], prefix='Тип_занятости')
data_new.drop(columns=['Подходящих тип занятости.name'], inplace=True)
```

```python
data_new = pd.get_dummies(data_new, columns=['Возможность переезда.type.id'], prefix='Возможность_переезда')
data_new.drop(columns=['Возможность переезда.type.name'], inplace=True)

data_new = pd.get_dummies(data_new, columns=['Подходящий график работы.id'], prefix='Подходящий_график_работы')
data_new.drop(columns=['Подходящий график работы.name'], inplace=True)

data_new.drop(columns=['Желательное время в пути до работы.id'], inplace=True)
data_new.drop(columns=['Желательное время в пути до работы.name'], inplace=True)

data_new = pd.get_dummies(data_new, columns=['Готовность к командировкам.id'], prefix='Командировки')
data_new.drop(columns=['Готовность к командировкам.name'], inplace=True)

```

```python
stage_mapping = {
    'response': 'response',
    'phone_interview': 'phone_interview',
    'interview': 'interview',
    'offer': 'hired',
    'hired': 'hired',
    'consider': 'consider',
    'discard_by_applicant': 'discard_other',
    'discard_no_interaction': 'discard_other',
    'discard_to_other_vacancy': 'discard_other',
    'discard_vacancy_closed': 'discard_other',
    'discard_by_employer': 'discard_by_employer'
}

data_new['Информация об этапе воронки найма'] = data_new['Информация об этапе воронки найма.Информация о состоянии.id'].map(stage_mapping)

print(data_new['Информация об этапе воронки найма'].value_counts())
```

```python
print(data_new['Информация об этапе воронки найма.Информация о состоянии.name'].value_counts())
```
