```python
import pandas as pd
import re

experience_mapping = {
    'без опыта': 0,
    'нет опыта': 0,
    'с 16 лет': 0,
    'не нужен': 0,
    'студентов на подработку': 0,
    'без опыта работы': 0,
    'без опыта работы': 0,
    '16 лет': 0,
    'подроботк': 0,
}

def convert_experience_to_numeric(experience):
    if isinstance(experience, str):
        experience = experience.lower()
        for key, value in experience_mapping.items():
            if key in experience:
                return value
    return None

df['Опыт работы числовое'] = df['Опыт работы'].apply(convert_experience_to_numeric)

print(df[['Опыт работы', 'Опыт работы числовое']].dropna())

```
